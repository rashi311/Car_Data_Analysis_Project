```python
#Start by importing all important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the csv file and assign it to a variable .
df=pd.read_csv('Cars.csv')
print(df.head(5))
```

```python
#Display shape of dataframe
print(df.shape)
```

```python
#Print all columns of dataframe.Return an array containing names of all the columns.
column_names=df.columns
print(column_names)
```

```python
#Set the 'name' column as the index of dataframe
df=df.set_index('name')
```

```python
df
```

```python
#Print a list of all the unique mpg values
unique_values=df['mpg'].unique().tolist()
print(unique_values)
```

```python
#Create a column which contains the horsepower divided by weightas its metric and make this new column the index.
df['hp_per_weight']=df['horsepower']/df['weight']
df.set_index('hp_per_weight')
```

```python
df = df.reset_index() 
df = df.set_index('name')

 
```

```python
df
```

```python
#What is name of car that has the highest horsepower?
highest_car_hp=df['horsepower'].max()
print(highest_car_hp)
```

```python
#How many cars have mpg â‰¥ 35?
count = (df['mpg'] >= 35).sum()
print(count)
```

```python
#What is the most common origin for cars with horsepower > 100 and weight < 3000?
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()
print(most_common_origin)
```

```python
#What is the mean acceleration of cars from Japan? (rounded to 2 decimals)
mean_acceleration=df['acceleration'].mean()
print(round(mean_acceleration))
```

```python
#Which year had the highest average mpg?
mpg_each_year=df.groupby('model_year')['mpg']
avg_mpg_each_year=mpg_each_year.mean()

highest_avg_mpg_year=avg_mpg_each_year.idxmax()
print(highest_avg_mpg_year)
```

```python
#Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

median_mpg=df['mpg'].median()
car_df = df[df['mpg']>median_mpg]
max_ratio=car_df['hp_per_weight'].max()
best_cars = car_df[car_df['hp_per_weight'] == max_ratio]
print(best_cars)
```

```python
#Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
pivot_table = avg_mpg.pivot(index='model_year', columns='origin', values='mpg')
pivot_table.plot(marker='o', figsize=(10, 6))
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

sns.lineplot(data=df, x='model_year', y='mpg', hue='origin', estimator='mean', marker='o')

# Customize
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''Create a Seaborn scatterplot (or PairGrid) where:
X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.
'''
sns.scatterplot(data=df,x="horsepower",y='weight',hue='origin',size='mpg',hue_order=['japan','europe','usa'])
plt.title('Horsepower vs Weight of Cars by Origin and MPG Size', fontsize=14)
plt.xlabel('horsepower')
plt.ylabel('weight')
plt.show()
```

```python
#Identify car names that appear in more than one model_year.
carnames_year_count=df.groupby('name')['model_year'].nunique()
duplicate_names = carnames_year_count[carnames_year_count > 1].index.tolist()
for name in duplicate_names:
    print(name)
```

```python
#For each such name, compute the standard deviation of mpg across years.
df=df.reset_index()

carnames_year_count=df.groupby('name')['model_year'].nunique()
duplicate_names = carnames_year_count[carnames_year_count > 1].index
duplicate_cars_df = df[df['name'].isin(duplicate_names)]
mpg_std_by_name = duplicate_cars_df.groupby('name')['mpg'].std()
print(mpg_std_by_name.sort_values(ascending=False))
```

```python
#Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0
name_year_counts = df.groupby('name')['model_year'].nunique()
duplicate_names = name_year_counts[name_year_counts >= 2].index
duplicate_cars_df = df[df['name'].isin(duplicate_names)]
mpg_std_by_name = duplicate_cars_df.groupby('name')['mpg'].std()
low_variation_cars = mpg_std_by_name[mpg_std_by_name < 1.0]
min_std = low_variation_cars.min()
cars_with_lowest_variation = low_variation_cars[low_variation_cars == min_std].index.tolist()


for car in cars_with_lowest_variation:
    print(car)
```

```python
#Report the model name(s), number of appearances, and the average mpg.
name_year_counts = df.groupby('name')['model_year'].nunique()
duplicate_names = name_year_counts[name_year_counts >= 2].index
duplicate_cars_df = df[df['name'].isin(duplicate_names)]
mpg_std_by_name = duplicate_cars_df.groupby('name')['mpg'].std()
low_variation_cars = mpg_std_by_name[mpg_std_by_name < 1.0]
min_std = low_variation_cars.min()
cars_with_lowest_variation = low_variation_cars[low_variation_cars == min_std].index.tolist()

result = duplicate_cars_df[duplicate_cars_df['name'].isin(cars_with_lowest_variation)].groupby('name').agg(
    appearances=('model_year', 'count'),
    average_mpg=('mpg', 'mean')
).reset_index()

print(result)
```
